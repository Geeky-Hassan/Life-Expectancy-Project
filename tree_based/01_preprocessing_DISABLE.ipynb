```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
my_data = pd.read_csv("./dataset/life_expectancy.csv")
my_data.head()
```

```python
orig_cols = list(my_data.columns)
new_cols = []
for col in orig_cols:
    new_cols.append(col.strip().replace('  ', ' ').replace(' ', '_').replace('-', '_').lower())
    
my_data.columns = new_cols
my_data.rename(columns={'thinness_1_19_years':'thinness_10_19_years'}, inplace=True)
```

```python
continent = pd.read_csv("./dataset/countryContinent.csv", encoding='ISO-8859-1')
continent.head()
```

```python
country2continent = dict()
for country in continent['country']:
  country2continent[country] = continent[continent['country']==country]['continent'].values[0]

update = {'Republic of Korea' : 'Asia', 'Republic of Moldova' : 'Europe' ,
          'Democratic Republic of the Congo' : 'Africa', 'Czechia' : 'Europe', 
          'United Republic of Tanzania' : 'Africa', 'The former Yugoslav republic of Macedonia' : 'Europe', 
          "Democratic People's Republic of Korea" : 'Asia'}

country2continent.update(update)
```

```python
my_data['continent'] = my_data['country'].copy().replace(country2continent)
my_data.continent.isna().sum()
```

```python
my_data['continent'].unique()
```

```python
def remove_whitespace(x):
    try:
        x = "".join(x.split())
    except:
        pass
    return x

my_data.country = my_data.country.apply(remove_whitespace)
```

```python
my_data = pd.get_dummies(my_data, columns=['status', 'continent'])

X = my_data.drop(columns='life_expectancy', axis=1)
y = my_data['life_expectancy']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
num_col = list(X_train.select_dtypes(include=['int64', 'float64']).columns)
num_col.remove('adult_mortality')

X_train.loc[:, num_col] = X_train.loc[:, num_col].fillna(X_train[num_col].mean())
X_train.reset_index(drop=True)

X_test.loc[:, num_col] = X_test.loc[:, num_col].fillna(X_train[num_col].mean())
X_test.reset_index(drop=True)

my_data.loc[:, num_col] = my_data.loc[:, num_col].fillna(X_train[num_col].mean())
my_data.reset_index(drop=True)
 
my_data.shape
```

```python
#from google.colab import files

my_data.to_csv('./dataset/my_data.csv', encoding = 'utf-8-sig', index=False) 
#files.download('my_data.csv')

X_train.to_csv('./dataset/X_train.csv', encoding = 'utf-8-sig', index=False) 
#files.download('X_train.csv')

X_test.to_csv('./dataset/X_test.csv', encoding = 'utf-8-sig', index=False) 
#files.download('X_test.csv')

y_train.to_csv('./dataset/y_train.csv', encoding = 'utf-8-sig', index=False) 
#files.download('y_train.csv')

y_test.to_csv('./dataset/y_test.csv', encoding = 'utf-8-sig', index=False) 
#files.download('y_test.csv')
```
